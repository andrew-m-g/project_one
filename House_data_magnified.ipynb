```python
df = pd.read_csv("City_Zhvi_AllHomes.csv")

columns_2005 = df.filter(like='2005')

average_2005 = columns_2005.mean(axis=1)


df['Average_2005'] = average_2005


state_mean_price = df.groupby('State')['Average_2005'].mean()


plt.figure(figsize=(10, 6))
state_mean_price.plot(kind='bar')
plt.title('Average House Price by State in January 2005')
plt.xlabel('State')
plt.ylabel('Average House Price')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python

columns_2012 = df.filter(like='2012')


average_2012 = columns_2012.mean(axis=1)

# Add the average values as a new column named 'Average_2012' to the DataFrame
df['Average_2012'] = average_2012

# Print the DataFrame with the new column
print(df)
```

```python
df = pd.read_csv('City_Zhvi_AllHomes.csv')


indianapolis_data = df[(df['RegionName'] == 'Indianapolis') & (df['State'] == 'IN')]
time_period_data = indianapolis_data.loc[:, '2008-03-31':'2008-07-31']


time_period_data = time_period_data.T


time_period_data.reset_index(inplace=True)


time_period_data.columns = ['Date', 'Housing Price']


import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(time_period_data['Date'], time_period_data['Housing Price'], marker='o')
plt.title('Housing Prices in Indianapolis (March 2008 - July 2008)')
plt.xlabel('Date')
plt.ylabel('Housing Price ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.savefig('indianapolis_housing_prices.png')

plt.show()

```
